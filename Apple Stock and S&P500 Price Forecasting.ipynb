```python
import pandas as pd
import yfinance as yf
from prophet import Prophet
import matplotlib.pyplot as plt

# Downloads Apple stock data
aapl = yf.download('AAPL', start='2017-01-01', end='2025-06-05')

# Uses only the 'Close' price and reset index
aapl = aapl[['Close']].reset_index()

# Renames columns to fit Prophet's requirements
aapl.columns = ['ds', 'y']

# Ensures correct data types
aapl['ds'] = pd.to_datetime(aapl['ds'])        # Makes sure 'ds' is datetime
aapl['y'] = pd.to_numeric(aapl['y'], errors='coerce')  # Ensures 'y' is float

# Drops any rows with missing values
aapl.dropna(inplace=True)

# Fits the model
model = Prophet(daily_seasonality=True)
model.fit(aapl)

# Forecasts the future price
future = model.make_future_dataframe(periods=365)
forecast = model.predict(future)

# Plots the forecast
fig1 = model.plot(forecast)
plt.title("Apple Price Forecast")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.grid(True)
plt.show()
```

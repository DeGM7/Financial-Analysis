# __Bitcoin and Ethereum Price Forecasting__

## ___Bitcoin Price Forecasting___

```python
import pandas as pd
import yfinance as yf
from prophet import Prophet
import matplotlib.pyplot as plt

# Downloads Bitcoin data
btc = yf.download('BTC-USD', start='2017-01-01', end='2025-06-05')

# Uses only the 'Close' price and reset index
btc = btc[['Close']].reset_index()

# Renames columns to fit Prophet's requirements
btc.columns = ['ds', 'y']

# Ensures correct data types
btc['ds'] = pd.to_datetime(btc['ds'])        # Makes sure 'ds' is datetime
btc['y'] = pd.to_numeric(btc['y'], errors='coerce')  # Ensures 'y' is float

# Drops any rows with missing values
btc.dropna(inplace=True)

# Fits the model
model = Prophet(daily_seasonality=True)
model.fit(btc)

# Forecasts the future price
future = model.make_future_dataframe(periods=365)
forecast = model.predict(future)

# Plots the forecast
fig1 = model.plot(forecast)
plt.title("Bitcoin Price Forecast")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.grid(True)
plt.show()
```

## ___Ethereum Price Forecasting___
